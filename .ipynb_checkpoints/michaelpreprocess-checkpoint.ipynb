```python
df = pd.read_csv('../UNSW_NB15_training-set.csv', index_col = False).drop('id', axis= 1)
df
```

```python
TCP = df[df['proto'] == 'tcp']
TCP
```

```python
TCP_NON_SERVICE = TCP[TCP['service'] == '-']
TCP_NON_SERVICE
```

```python
TCP_NON_SERVICE.loc[TCP_NON_SERVICE['state'] == 'FIN', 'state'] = 0
TCP_NON_SERVICE.loc[TCP_NON_SERVICE['state'] == 'CON', 'state'] = 1
TCP_NON_SERVICE.loc[TCP_NON_SERVICE['state'] == 'REQ', 'state'] = 2
```

```python
TCP_NON_SERVICE['state'].value_counts()
```

```python
ACC = TCP_NON_SERVICE[TCP_NON_SERVICE['state'] == 'ACC'].index
CLO = TCP_NON_SERVICE[TCP_NON_SERVICE['state'] == 'CLO'].index
RST = TCP_NON_SERVICE[TCP_NON_SERVICE['state'] == 'RST'].index
TCP_NON_SERVICE = TCP_NON_SERVICE.drop(ACC)
TCP_NON_SERVICE = TCP_NON_SERVICE.drop(CLO)
TCP_NON_SERVICE = TCP_NON_SERVICE.drop(RST)
```

```python
TCP_NON_SERVICE = TCP_NON_SERVICE.drop(
    ['service', 
     'proto', 
     'ct_ftp_cmd', 
     'is_sm_ips_ports', 
     'ct_flw_http_mthd', 
     'is_ftp_login'], axis=1)
TCP_NON_SERVICE
```

```python
NORMAL = TCP_NON_SERVICE[TCP_NON_SERVICE['attack_cat'] == 'Normal']
ANOMALY = TCP_NON_SERVICE[TCP_NON_SERVICE['attack_cat'] != 'Normal']
```

```python
NORMAL = NORMAL.drop(['attack_cat', 'label'], axis = 1)
ANOMALY = ANOMALY.drop(['attack_cat', 'label'], axis = 1)
```
